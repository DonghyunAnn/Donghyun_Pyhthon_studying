numpy.ndarray

- 동일한 자료형을 가지는 값들이 배열 형태로 존재함
- N 차원 형태로 구성이 가능하다.
- 각 값들은 양의 정수로 색인(index)가 부여되어 있다.
- Numpy에서 차원(Dimension)을 rank, axis라고 부르기도 한다.
- ndarray를 줄여서 array로 표현한다.


````
list=[1,2,3]
array1 = np.array(list)
print( type(array1) ) #타입은 함수니까 괄호안에 파라미터를 넣어주는것
print( array1.shape ) #배열의 크기확인

array2 = np.array( [ [1,2,3],
                    [4,5,6] ] ) #np 안에 들어있는 함수를 사용
print( type(array2) )
print( array2.shape ) # 객체= 속성!! 이니까 .속성나오는것 나오는것
array3 = np.array([[1,2,3]])
print( type(array3) )
print( array3.shape )
````

ndarray.shape

해당 array의 크기를 알 수 있다.

- ndarray의 차원과 크기를 튜플(tuple) 형태로 출력

- shape 을 확인함으로써 몇개의 데이터가 있는지, 몇 차원으로 존재하는지 등을 확인할 수 있다.

- 위에서 array1.shape의 결과는 (3,) 으로써, 1차원의 데이터이며 총 3라는 크기를 갖고 있음을 알 수 있다.

- array2.shape의 결과는 (2,3) 으로써, 2차원의 데이터이며 2 * 3 크기를 갖고 있는 array 이다.


````
# array 차원(Dimension) 확인
array1.ndim, array2.ndim, array3.ndim
````

````
# array의 전체 요소 개수 확인
array1.size, array2.size, array3.size
````

numpy에서 사용되는 자료형

- 부호가 있는 정수 int(8, 16, 32, 64)
- 부호가 없는 정수 uint(8 ,16, 32, 54)
- 실수 float(16, 32, 64, 128)
- 복소수 complex(64, 128, 256)
- 불리언 bool
- 문자열 string_
- 파이썬 오프젝트 object
- 유니코드 unicode_


````
# ndarray내의 데이터 타입 확인
list1=[1,2,3]
print(type(list1))

array1= np.array(list)
print(type(array1))
print(array1)
print(array1.dtype)
````

````
# 서로 다른 데이터 타입을 갖을 때
list2=[1,2,'test']
array2 = np.array(list2)
print( array2, array2.dtype )

list3= [1,2,3.0]
array3 = np.array(list3)
print(array3, array3.dtype)
````


- 리스트와 다르게 ndarray는 같은 데이터 타입만 가능.
- 서로 다른 데이터 타입이 섞여 있을 경우 큰 데이터 타입으로 형 변환을 일괄 적용

````
# 데이터 타입 변경하기
array_int = np.array([1,2,3])
print(array_int, array_int.dtype)

array_float=array_int.astype(float)
print(array_float, array_float.dtype)

array_int32 = array_int.astype('int32')
print(array_int32, array_int32.dtype)

array_float2=np.array([1.2,2.1,3.1])
array_int2 = array_float2.astype('int32')
print(array_float2, array_float2.dtype)
print(array_int2, array_int2.dtype)
````

````
# 타입 지정하여 생성
array_int = np.array( [1.1,2.1,3.1] , dtype=np.int64 )
print( array_int, array_int.dtype)
````

### ndarray 차원과 크기 변경

````
array1 = np.arange(10)
print(array1)

array2 = array1.reshape(2,5)
print(array2)

array3 = array1.reshape(5,2)
print(array3)
````

````
# 자동변환 할 때는 -1을 넣어주면 됨, 자동으로 호환됨
array1 = np.arange(10)
print(array1)

array2 = array1.reshape(-1,5)
print(array2)

array3 = array1.reshape(-1,2)
print(array3)
````

````
# ndarray 형태를 통일
# 원본 ndarray가 어떤 형태라도 2차원이고 1개의 컬럼을 갖는 ndarray로 변환
array1 = np.arange(8)
array3d = array1.reshape( (2,2,2) )
print( array3d.tolist() )

array5 = array3d.reshape(-1,1)
print(array5.tolist(), array5.shape)

array6= array1.reshape(-1,1)
print(array6.tolist(),array6.shape)
````

### ndarray 인덱싱 & 슬라이싱
- 특정 데이터 추출
- 슬라이싱(Slicing) : 연속된 데이터 추출
- 팬시 인덱싱(Fancy Indexing) : 리스트 또는 ndarray 형태의 인덱싱 집합을 이용하여 여러 데이터를 추출
- 불린 인덱싱(Boolean Indexing) : True/False 값 인덱싱 집합을 이용하여 해당 인덱스 위치에 있는 데이터를 추출

````
# 1부터 9까지의 1차원 ndarray 생성
array = np.arange(start = 1, stop= 10)
print(array)

v = array[2]
print('value: ', v)
print(type(v))
````

````
# 인덱스를 이용한 데이터 수정
array[0] = 9 
array[8] = 0
array
````

````
array = np.array( [[1,2,3],
                   [4,5,6]] )

print(array[0])
print(array[0][0])
print(array[0,0])
````

````
array1d = np.arange(1,10)
array2d = array1d.reshape(3,3)

print(array2d)
print("row = 0, col = 0", array2d[0,0])
print("row = 0, col = 1", array2d[0,1])
print("row = 0, col = 0", array2d[0,2])

````

슬라이싱
- ':' 기호 앞에 시작 인덱스를 생략하면 맨 처음 인덱스인 0으로 간주
- ':' 기호 뒤에 종료 인덱스를 생략하면 맨 마지막 인덱스로 간주
- ':' 기호 앞/뒤에 시작/종료 인덱스를 생략하면 맨 처음/마지막 인덱스로 간주



````
# 1차원 슬라이싱
arr = np.arange(1,10)
arr[0:3], arr[:3], arr[3:], arr[:]
````
