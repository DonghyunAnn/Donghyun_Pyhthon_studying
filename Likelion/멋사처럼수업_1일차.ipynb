# 인공지능 개요

* 문제는 지적인 기계가 어떤 감정을 가질 수 있느냐가 아니라 기계가 아무런 감정 없이 지능을 가질 수 있느냐 하는 것이다. - 마빈 민스키(1986년)/전 MIT 교수

* 특이한 인간의 특성은 결코 기계에 의해 모방될 수 없다는 말은 일반적으로 위안이 된다. 나는 그런 위안을 줄 수 없다. 그런 한계는 정해질 수 없다고 믿기 때문이다. - 앨런튜링(1951년)/영국 수학자

# 인공지능이란?
* 인간의 지능(학습능력,추론능력,지각능력 등)을 인공적으로 구현한 컴퓨터 프로그램 또는 이를 포함한 컴퓨터 시스템
  * ex) 게임 AI: 어떤 의도가 있는 것처럼 행동하는 특징 (일정 범위에서만 움직임, 플레이어를 인식해 추적, 장애물을 파해 추적)
  * ex) 청소로봇, 자율주행, 스마트 스피커

# 인공지능의 보편적 기준
* 지능이 있는 것 처럼 보인다
  * 인간의 감각에 의존하기 때문에 기준이 애매힘
* 학습, 인식, 판단 능력이 있다
  * 추측 변환 이나 오차 수정 도 인공지능에 포함, 단순한 사양의 게임AI는 인공지능에 포함되지 않음


# 인공지능의 시작
*  주요 주제
 * 컴퓨터가 인간의 법칙을 이해하는 것
 * 인간의 뇌처럼 정보를 전달할 것
 * 이것들은 고도로 복잡한 계산 처리 기능이 필요함

* 프로그램의 원조= 튜링머신
  * 하나의 테이프에 명령을 쓰고 기계가 그것을 실행, 프로그램으로 정보처리를 하는 기초이론

* 컴퓨터 원형 = 노이만형 컴퓨터
  * 프로그램을 데이터로 기억하고 그 데이터를 숱서대로 읽고 실행하는 컴퓨터라는 개념을 만듬, 지금의 범용 컴퓨터의 원형

* 뇌 신경세로 재현 = 인공뉴련 -> 딥러닝
  * 인간 뇌의 신경세포를 모방해 만듬, 여러 개의 입력층에서 받아들인 정보를 처리하고 출력층으로 보냄

* 다트머스 회의가 개최되며 다음 분기로 넘어감

# 인공지능 접근법
* 기호주의(Symbolism)
  * 인간의 지능과 지식을 기호화해 메뉴얼화하는 접근법
  * 문제에 대응하기 위한 메뉴열이 있고 그 메뉴얼에 따르는 컴퓨터가 있으면 됨
  * 논리로 해결할 수 있는 문제, 메뉴얼화할 수 있는 분야에 강함, 반대 분야에서는 약함

* 연결주의(Connectionism)
  * 인간의 뇌 신경 네트워크의 재현을 목표로 하는 접근법
  * 인공지능 학습에 필요한 좋은 경험(정보)이 있으면 됨, 직접 또는 과거 통계 데이터를 사용
  *인간이 말로 설명할 수 없는 문제 또는 경험으로 배우는 것이 빠른 분야에 강함, 어떤식으로 생각하는지 인간도 모름

# 초기 인공지능의 한계

* 초개 챗봇에서 단순히 기계가 언어를 이해한 것인지에 대하여 의문이 생김 -> 자연어 처리의 발전

* 인간이 잘하는 것을 인공지능은 어려워 한다. - "모라베의 역설"
  * 논리문제 (AI)
  * 감각문제 (인간)

* 프레임 문제
  * 프레임 범위 안에서만 작동

* 조합 폭발
  * 프레임을 크게 만들면 범위는 넓어지지만 학습량이 폭발적으로 늘어나 처리중지

# 인공지능의 성장
* 결국 인공지능은 '지식'을 손에 넣었고 기계학습으로 성장했다.

## 1차 인공지능 봄의 종언
* 1차 인공지능 붐이 끝났던 원인
  * 처음 기대가 너무 컸다.
  * 인공지능이 세상을 몰랐다.
  * 인공지능을 다루는 컴퓨터의 성능이 너무 낮았다

* 인공지능에게 세상을 이해시키기위해 지식을 가르쳤음 = 정보들의 관련성을 알려줌= 지식표현

* 인공지능 사회 진출
  * 컴퓨터 성능 향상(지식저장)
  * 전문가 시스템 탄생(문답)
  * 의료 금융 분야로 확대
  * 하지만 지식의 폭발과 같은 문제가 발생

## 2차 인공지능 붐
* 기계학습의 도래
  * 학습을 기계에게 맡김, 인공지능의 학습효율을 올리기 위해
  * 과거의 경험이난 통계 데이터를 모아놓고 인공지능이 직접 지식을 습득하게 함
  * 지도학습 - Q&A
  * 비지도학습 - Q only
  * 강화학습 - 보상 UP : 문제를 푸는 학습 방향성만 제시

## 2차 인공지능 봄의 종언
* 정보(데이터) 부족
* 자율주행차와 단어의 발음을 학습하는 인공지능 등장
* 그러나 어떤 업무에서도 인간을 대신할 수준에 미달
* 기계학습에 필요한 중요 정보가 압도적으로 부족

## 인공지능의 혁신
* 딥러닝이라는 기계학습의 등장
