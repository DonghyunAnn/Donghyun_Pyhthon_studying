numpy.ndarray

- 동일한 자료형을 가지는 값들이 배열 형태로 존재함
- N 차원 형태로 구성이 가능하다.
- 각 값들은 양의 정수로 색인(index)가 부여되어 있다.
- Numpy에서 차원(Dimension)을 rank, axis라고 부르기도 한다.
- ndarray를 줄여서 array로 표현한다.


````
list=[1,2,3]
array1 = np.array(list)
print( type(array1) ) #타입은 함수니까 괄호안에 파라미터를 넣어주는것
print( array1.shape ) #배열의 크기확인

array2 = np.array( [ [1,2,3],
                    [4,5,6] ] ) #np 안에 들어있는 함수를 사용
print( type(array2) )
print( array2.shape ) # 객체= 속성!! 이니까 .속성나오는것 나오는것
array3 = np.array([[1,2,3]])
print( type(array3) )
print( array3.shape )
````

ndarray.shape

해당 array의 크기를 알 수 있다.

- ndarray의 차원과 크기를 튜플(tuple) 형태로 출력

- shape 을 확인함으로써 몇개의 데이터가 있는지, 몇 차원으로 존재하는지 등을 확인할 수 있다.

- 위에서 array1.shape의 결과는 (3,) 으로써, 1차원의 데이터이며 총 3라는 크기를 갖고 있음을 알 수 있다.

- array2.shape의 결과는 (2,3) 으로써, 2차원의 데이터이며 2 * 3 크기를 갖고 있는 array 이다.

#### Numpy 예제 04

영화평점 데이터 분석

````
import numpy as np

# 사용자 아이디, 영화 아이디, 평점, ?
data = np.loadtxt(drive_path + "ratings_small.txt", delimiter = "::", dtype = np.int64)
data
````

##### 데이터 분석하기
```
print(data.ndim)
print(data.shape)
print(data.size)
```

````
# 행(평점 작성개수) 300004개 열(데이터의 종류) 4개
# 300004 * 4 = 1200016
print(data.ndim)
print(data.shape)
print(data.size)
````

##### 전체 평점 평균 구하기
1. 평점 데이터만 추출 (인덱싱, 슬라이싱)
2. 평균 구하기 (mean)

````
# 전체 평점 평균 평점
data_grade = data[:,2]
data_grade.mean()
````

##### 각 사용자별 평점 평균 구하기
1번 사용자의 평균 평점 구하기
1. 1번 사용자의 데이터만 Boolean Indexing 추출
2. 1번 사용자의 데이터에서 평점만 추출 (인덱싱, 슬라이싱)
3. 평균 구하기

참고
```
# 중복된 성분을 제외한 array를 반환
np.unique( ndarray )
```

````
#1번 사용자의 데이터만 Boolean Indexing 추출
rate = data[data[:,0] == 1]
#1번 사용자의 데이터에서 평점만 추출 (인덱싱, 슬라이싱)
rate[:,2]
````

````
# 1번 사용자 평점 평균
np.mean(rate[:,2])
````

````
# 각 사용자별 평점 평균
lst=[]
for i in range(1,1780):
  id = data[data[:,0] == i]
  rate_mean = np.mean(id[:,2])
  print("id : {0} \t star rate: {1}".format(i,rate_mean))
````

````
user_id_only = np.unique(data[:, 0])
lst=[]
for i in user_id_only:
  id = data[data[:,0] == i]
  rate_mean = np.mean(id[:,2])
  lst.append([i,rate_mean])
````

##### 각 사용자 별 평균 평점이 4점 이상인 사용자 구하기
- 리스트는 Boolean Indexing이 안되니 배열로 변환해야 인덱싱, 슬라이싱 가능

````
# 각 사용자 별 평균 평점이 4점 이상인 사용자 구하기
score_array = np.array(lst)
boolean_index = score_array[:,1] >= 4
score_array[boolean_index][:,0].astype(int)
````

##### CSV 파일로 저장
- np.savetxt() 이용

````
# csv 저장
np.savetxt(drive_path + 'user_id_mean.csv', lst, delimiter=', ', fmt='%.3f')
````

````
# 열을 기준으로 소팅
x = np.array([[8,4,11,5],
              [2,10,7,3],
              [9,12,1,6]])
x.sort(0)
x
````
