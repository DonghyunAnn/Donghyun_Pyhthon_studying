### 집합(set)
- 중복을 허용하지 않는다.
- 순서가 없다(Unordered). 순서가 없는 자료형이라 인덱싱을 지원하지 않는다.

※ 중복을 허용하지 않는 set의 특징은 자료형의 중복을 제거하기 위한 필터 역할로 종종 사용하기도 한다.


````
# 리스트와 튜플을 입력값으로 받을 수 있음
s1= set([1,2,3,3])
s1
````

````
# 교집합, 합집합, 차집합
s1 = set([1,2,3,4,5,6])
s2= set([4,5,6,7,8,9])

print(s1 & s2)
print(s1.intersection(s2))

print(s1 | s2)
print(s1.union(s2))

print(s1 - s2)

print(s1.difference(s2))
````

````
# 집합에서 요소를 더할 때
s1 = set([1,2,3])
s1.add(4)
print(s1)
# 집합에서 여러개를 더할 때
s1.update([4,5,6,7,8])
print(s1)
# 집합에서 없앨 때 
s1.remove(2)
print(s1)
````

매개변수와 인수

매개변수(parameter)와 인수(arguments)는 혼용해서 사용되는 헷갈리는 용어이므로 잘 기억해 두자. 매개변수는 함수에 입력으로 전달된 값을 받는 변수를 의미하고 인수는 함수를 호출할 때 전달하는 입력값을 의미한다.

````
def add(a, b):  # a, b는 매개변수
    return a+b

print(add(3, 4))  # 3, 4는 인수
````

````
# 매개변수 지정하여 호출
result = add(b=3,a=4) #위치변경 가능 
print(result)
````

````
# 입력값이 몇 개가 될지 모를 때
def add_many(*args):
  result=0
  for i in args:
    result += i
  return result

add_many(1,2,3,4,5,7)
````

````
def calc_mul(oper,*args):
  result = 0
  if oper == 'add':
    for i in args:
      result += i

  if oper =='mul':
    result = 1
    for i in args:
      result *= i

  return result

print(calc_mul('add',1,2,3,4,5))
print(calc_mul('mul',1,2,3,4,5))
````

````
# 키워드 파라미터 kwargs
# 매개변수 kwargs는 딕셔너리가 되고 모든 key=value 형태의 결과값이 그 딕셔너리에 저장된다.
def print_kwargs(**kwargs):
  print(kwargs)

print_kwargs(a=1)
print_kwargs(name='foo',age = 3)
````

````
#따로 해보기
#최대공약수
def LargeSameDiv(a,b):
  lst1=[]
  lst2=[]
  for i in range(1,a):
    if a % i == 0:
      lst1.append(i)
  lst1.append(a)
  for i in range(1,b):
    if b % i == 0:
      lst2.append(i)
  lst2.append(b)

  lst1=set(lst1) ; lst2=set(lst2)  
  lst=lst1&lst2
  
  return max(lst)

LargeSameDiv(3,10)
````

````
#따로 해보기
#최소공배수
def LargeSameMul(a,b):
  for i in range(1,a+1):
    if int(LargeSameDiv(a,b)) * i == a:
      v1=i
  for j in range(1,b+1):
    if int(LargeSameDiv(a,b)) * j == b:
      v2=j
  return LargeSameDiv(a,b) * v1 * v2
LargeSameMul(5,7)
````

객체와 인스턴스의 차이는
> 인스턴스라는 말은 특정 객체(a)가 어떤 클래스(Cookie)의 객체인지를 관계 위주로 설명할 때 사용한다. "a는 인스턴스"보다는 "a는 객체"라는 표현이 어울리며 "a는 Cookie의 객체"보다는 "a는 Cookie의 인스턴스"라는 표현이 훨씬 잘 어울린다. (점프 투 파이썬)


파이썬 클래스의 구조
```
class 클래스명:
    def __init__(self, param):
        .......
```
```
class 클래스명:
    def __new__(cls):
        .......
    def __init__(self, param):
        .......
    def 메소드명(self, param):
        .......
```

````
#함수
def calculate(a,b,c):
  if c == "+": return a+b
  elif c == "-": return a-b
  elif c == "*": return a*b
  elif c == "/": return a/b
  else: print("연산자 입력 오류")
  
calculate(2,3,'+')
calculate(3,1,'-')
calculate(8,2,'*')
calculate(8,4,'/')
calculate(8,4,'^')
````

````
#클래스
class Calculator:
  def __init__(self,v1,v2):
    self.v1 = v1
    self.v2 = v2
  
  def sum(self):
    return self.v1 + self.v2
  def divide(self):
    return self.v1 / self.v2
  def subtract(self):
    return self.v1 - self.v2
  def product(self):
    return self.v1 * self.v2
  
calc = Calculator(4,2)
calc.sum()
````

````
# 연습문제
import random

class Account:
  # 클래스 변수
  account_count = 0

  def __init__(self, name, balance):
    self.deposit_count = 0  # 입금횟수 초기화
    self.deposit_log = []   # 입금내역
    self.withdraw_log = []  # 출금내역

    # 빠뜨린 속성 추가 구현
    self.name=name
    self.balance = balance    
    self.bank = "SC은행"

    # 랜덤하게 개좌번호 생성을 구현
    num1=random.randint(0,999)
    num1=str(num1).zfill(3)
    num2=random.randint(0,99)
    num2=str(num1).zfill(2)
    num3=random.randint(0,999999)
    num3=str(num1).zfill(6)
    
    
    self.account_number = num1 + '-' + num2 + '-' + num3
    Account.account_count += 1
  
  @classmethod
  def get_account_num(cls):
    print(cls.account_count)
  
  def deposit(self, amount):
    if amount > 1:
      self.deposit_log.append(amount)
      self.balance += amount

      self.deposit_count += 1
      if self.deposit_count % 5 == 0:   # 5, 10, 15 ...
        self.balance *= 1.01            # 이자
    
  def withdraw(self, amount):
    if self.balance > amount:
      self.withdraw_log.append(amount)
      self.balance -= amount
    
  def display_info(self):
    # 정보출력을 구현
    print("은행이름: ",self.bank)
    print("예금주: ",self.name)
    print("계좌번호: ",self.account_number)
    print("잔고: ",format(self.balance,','))
  
  def deposit_history(self):
    # 예금 히스토리 출력을 구현
    for i in range(len(self.deposit_log)):
          print(self.deposit_log[i])
  
  def withdraw_history(self):
    # 출금 히스토리 출력을 구현
    for i in range(len(self.withdraw_log)):
          print(self.withdraw_log[i])


kim = Account("김민수", 100)
print("은행이름:", kim.bank) 
print("계좌번호:", kim.account_number) 
print('-'*20)

kim = Account("김민수", 100)
print("총 계좌수:", Account.account_count)
lee = Account("이민수", 100)
print("총 계좌수:", Account.account_count)
print('-'*20)

kim = Account("김민수", 100)
lee = Account("이민수", 100)
Account.get_account_num()
print('-'*20)

kim = Account("김민수", 100)
kim.deposit(1)
print(kim.balance)
kim.deposit(100)
print(kim.balance)
print('-'*20)

p = Account("파이썬", 10000)
p.display_info()
print('-'*20)

kim = Account("김민수", 100)
kim.deposit(100)
kim.withdraw(90)
print(kim.balance)
print('-'*20)

p = Account("파이썬", 10000)
p.deposit(100)
p.deposit(200)
p.deposit(300)
p.deposit_history()
print('-'*20)

p.withdraw(100)
p.withdraw(200)
p.withdraw_history()
print('-'*20)
````


생성자로 객체생성을 호출받으면 먼저 __new__ 를 호출하여 객체를 생성할당하고, __new__ 메소드가 __init__메소드를 호출하여 객체에서 사용할 초기값들을 초기화하게됩니다.
