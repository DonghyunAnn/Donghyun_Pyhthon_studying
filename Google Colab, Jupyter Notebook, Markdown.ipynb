## 매직 명령어(magic commands)

* Jupyter Notebook 환경에서 파이썬 파일 저장 및 실행

```
%%writefile test.py
print("Hello Colab")
```

```
%run test.py
```

* 코드 실행 시간 측정

```
%%timeit?
```

```
#한줄
%timeit ls=[n**n for n in range(10000)]
```

```
#여러줄
%%timeit
ls=[]
for n in range(10000):
  ls.append(n**n)
```
